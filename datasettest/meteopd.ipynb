```python
import pandas as pd
import numpy as np
```

```python
url ="https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-min-temperatures.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.describe()
```

```python
dprow = df[df.duplicated()]
dprow
#no dupliccate
```

```python
dnull= df[df['Temp'].isna()]
dnull
#no na value
```

```python
# creer month et year a partir de date
#tri chronologique
# df = df[df['Date'].sort_values(ascending= True)]
df['Date'] = pd.to_datetime(df['Date'])
df['Year']= df['Date'].dt.year
df['day']= df['Date'].dt.day
df["Month"]= df['Date'].dt.month
df['Date'] =df["Date"].sort_values()
df.drop(columns='Date')
```

```python
#groupby et rolling average
df['rolligntemp7'] = df['Temp'].rolling(window=7).mean()
df
```

```python
df['rollingmonth']= df['Temp'].rolling(window=30).mean()
df
#rolling mean from the last month
```

```python
# mean by month and year
df['mean_month']= df.groupby(['Month', 'Year'])['Temp'].transform('mean')
df
```
